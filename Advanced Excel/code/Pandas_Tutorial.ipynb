VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,5))
sns.histplot(data=df, x='Total', bins=30, kde= True)
plt.title('Distribution of Total Stats')
plt.xlabel('Total Stats')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Legendary')
plt.title('Attack v/s Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend(title= 'Legendary', loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Type 1', order=df['Type 1'].value_counts().index, palette='viridis')
plt.title('Frequency of Pokemon Types (Type 1)')
plt.xlabel('Type 1')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Generation', y='Total', palette='coolwarm')
plt.title('Total Stats Distribution By Generation')
plt.xlabel('Generation')
plt.ylabel('Total Stats')
plt.show()
```

```python
legendary_counts = df['Legendary'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(legendary_counts, labels=legendary_counts.index, autopct='%1.1f%%', colors=['skyblue','lightcoral'])
plt.title('Proportion of The Legendary Pokemon')
plt.show()
```

```python
print(legendary_counts)
```

```python
type1_count = df['Type 1'].value_counts()

sns.set_palette("pastel")

plt.figure(figsize=(10,8))
plt.pie(type1_count, labels=type1_count.index, startangle=90, counterclock= False, autopct='%1.1f%%')
plt.title('Distribution of Type 1 Pokemon')
plt.axis('equal')
plt.show()
```

```python
generation_totals = df.groupby('Generation')['Total'].mean().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=generation_totals, x='Generation', y='Total', marker='o')
plt.xlabel('Generation')
plt.ylabel('Mean Total Stats')
plt.xticks(generation_totals['Generation'])
plt.grid(True)
plt.show
```

```python
numerical_columns = df[['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']]
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Pokemon Stats')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

x = df['Attack']
y = df['Defense']
z = df['Total']
colors = df['Generation']

sc = ax.scatter(x,y,z,c=colors, cmap='viridis', s=50)
ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
ax.set_zlabel('Total')
ax.set_title('3D Scatter Plot of Attack, Defense and Total Stats')
plt.colorbar(sc, label='Generation')
plt.show()
```

```python
sns.pairplot(df[['Total', 'HP','Attack','Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary']], hue='Legendary', palette='husl')
plt.suptitle('Pair Plot of Pokemon Stats with Legendary Status')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='Legendary', y='Speed', data =df, inner='quartile', palette='muted')
plt.title('Violin Plot of Speed by Legendary Status')
plt.xlabel('Legendary')
plt.ylabel('Speed')
plt.xticks(ticks=[0,1], labels=['Not Legendary', 'Legendary'])
plt.show()
```

FEATURE ENGINEERING

```python
# Create new features
# Total Stats without HP
df['Total_without_HP'] = df['Total'] - df['HP']
```

```python
df.head()
```

```python
# Normalising Features
from sklearn.preprocessing import MinMaxScaler

# Normalize attack, defence, speed
scalar= MinMaxScaler()

df[['Attack_scaled','Defence_scaled','Speed_scaled']] = scalar.fit_transform(df[['Attack', 'Defense','Speed']])
```
